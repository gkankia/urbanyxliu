```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
```

```python
place = "Linkoping, Sweden"
```

```python
tags = {"amenity": ["clinic", "dentist", "hospital", "pharmacy", "veterinary" ]}
gdf = ox.features_from_place(place, tags)
```

```python
# Convert the GeoDataFrame to a pandas DataFrame
#df = pd.DataFrame(gdf)

# Set the option to display all rows
#pd.set_option("display.max_rows", None)
gdf
```

```python
gdf.reset_index([0, 1])
new_gdf = gdf.reset_index(['element_type','osmid'])
new_gdf
```

```python
new_gdf.loc[new_gdf['element_type'] == "relation"]
```

```python
# Filter the GeoDataFrame to include only nodes, aka points
nodes_gdf = new_gdf[new_gdf['geometry'].type == 'Point']
nodes_gdf
```

```python
# Define the file path for saving the GeoPackage file
output_file = "/Users/giorgikankia/Documents/GitHub/Urbanyx LiU/data collection/osm/pois-health.geojson"

# Save the filtered GeoDataFrame to a GeoPackage file
nodes_gdf.to_file(output_file, driver='GeoJSON')
```
